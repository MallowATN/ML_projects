Logistic Regression - Gradient Descent with Pytorch

```python
import torch
```

```python
X = torch.tensor([1,2,3,4], dtype=torch.float32)
Y = torch.tensor([2,4,6,8], dtype=torch.float32)
```

```python
w = torch.tensor(0.0,dtype=torch.float32, requires_grad=True)
```

```python
# model prediction (forward pass)
def forward(x):
    return w * x
```

```python
# loss = neab squared error
def loss(y,y_predicted):
    return ((y_predicted-y)**2).mean()
```

```python
print(f'Prediction before training: f(5) = {forward(5):.3f}')
```

```python
learning_rate = 0.01
n_iters = 100
```

```python
for epoch in range(n_iters):
    # prediction = forward pass
    y_pred = forward(X)
    
    # loss
    l = loss(Y,y_pred)
    
    # Gradients = backward pass (UPDATED)
    l.backward() # calculate the gradient of our loss --- dl/dw
    
    # update weights (UPDATED)
#     w -= learning_rate * dw #This operation should not be a part of our computation... need to wrap with no_grad() function
    with torch.no_grad():
        w -= learning_rate * w.grad
    
    #zero gradients (ADDED) 
#     b/c when we call backwards, it will write our gradients and accumulate w.grad attribute
    w.grad.zero_()
    
        
    if epoch % 10 ==0:
        print(f'epoch {epoch+1}:w = {w:.3f}, loss = {l:.8f}')
print(f'Prediction after training: f(5) = {forward(5):.3f}')
```
