```python
# Drop our unique identifiers
df1 = df1.drop(['county','precinct'],axis=1)
```

```python
# Scale our dataset with StandardScaler
from sklearn.preprocessing import StandardScaler

X = df1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
 from scipy.cluster.vq import kmeans, vq

 ''' Choosing the number of Clusters'''
#create empty list for SSEs
sse_list = []

#calculate Sum of Squared Errors (SSE) for different numbers of clusters 
for k in range(2, 40):
    kmeans_instance = kmeans(X_scaled, k)
    sse_list.append(kmeans_instance[1])

#plot SSEs
plt.plot(range(2, 40), sse_list)
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors')
plt.title('Elbow Method')
plt.show()
```

```python
# Create kmeans instance 
kmeans_instance = kmeans(X_scaled, 12)
#Generate cluster labels for each point 
cluster_labels, _ = vq(X_scaled, kmeans_instance[0])

#Coordinates of cluster centers 
cluster_centers = kmeans_instance[0]

# # of datapoints that lie in each cluster around their respectiv mean
cluster_counts = np.bincount(cluster_labels)

# print(cluster_labels)
# print(cluster_centers)
```

```python
print(cluster_counts)
```

```python
df2 = df1.copy()
df2['cluster_labels'] = cluster_labels

#calculate mean values of each demographic feature for each cluster and group by the clusters' center
cluster_means = df2.groupby('cluster_labels').mean()

#print mean values for each cluster 
# print(cluster_means)

cluster_means_df = pd.DataFrame(cluster_means)
cluster_means_df
```

```python
# Plotting the kmeans result
cluster_means_df.plot.bar(figsize=(20, 8))
plt.title('Mean Values of Voting Features by Cluster')
plt.xlabel('Cluster Label')
plt.ylabel('Mean Value')

plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```
