#### Dietary Guidelines: https://www.dietaryguidelines.gov/resources/2020-2025-dietary-guidelines-online-materials/top-10-things-you-need-know-about-dietary

There are 4 overarching Guidelines in the 2020-2025 edition:  

* Follow a healthy dietary pattern at every life stage.  
* Customize and enjoy nutrient-dense food and beverage choices to reflect personal preferences, cultural traditions, and budgetary considerations.
* Focus on meeting food group needs with nutrient-dense foods and beverages, and stay within calorie limits.
* Limit foods and beverages higher in added sugars, saturated fat, and sodium, and limit alcoholic beverages.

There are Key Recommendations supporting 3 Guidelines, including quantitative recommendations on limits that are based on the body of science reviewed. The Guidelines recommend:

* Limiting added sugars* to less than 10% of calories per day for ages 2 and older and to avoid added sugars for infants and toddlers;
* Limiting saturated fat to less than 10% of calories per day starting at age 2;
* Limiting sodium intake to less than 2,300mg per day (or even less if younger than 14);

### Importing the packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataframe
df = pd.read_csv('cereal.csv')
df.head(3)
```

### Statistic

```python
# Information of the cereal dataframe
df.info()
```

```python
# Check for null values... There's no missing data
df.isnull().sum()
```

```python
# Descriptive Statistics of Cereal dataset
df.describe().T 
```

```python
# Assume the calorie intake of a normal healthy being is 2000.
calorie_intake_per_day = 2000

# how much grams of sugar/fat per calorie
grams_sugar_per_calorie = 4
grams_fat_per_calorie = 9

# dietary guidelines recommendation for sugar and fat.
limit_sugar = 0.1
limit_fat = 0.1

# How much grams of sugar/fat per calorie is allowed
sugar_allowed = int(calorie_intake_per_day * limit_sugar / grams_sugar_per_calorie)
fat_allowed = int(calorie_intake_per_day * limit_fat / grams_fat_per_calorie)

print(f'Sugar allowed: {sugar_allowed}g | fat allowed: {fat_allowed}g')
```

So if we compare the max value... (max value of sugar = 35g and max value of fat = 5g)... all cereals meet the requirements for the dietary guidelines.

```python
# Here we sort by the min value of our columns and see the top 5 candidates for the cereal under the guidelines. 
# (It could honestly be anything depending on the individual preferences, but for this problem, I'll stick with the guidelines)

def top_n_cereal(df, n, age):
    if age < 2:
        #Now according to the guidelines, it says 'avoid added sugars for infants and toddlers' and 'sodium intake should be EVEN LESS if younger than 14'.
        #I actually looked it up... and apparently high fibers contains phytates which binds with other nutrients and remove them from body. Bad for infants... which is why I also added fiber in the columns.
        return df.nsmallest(n=n, columns=['sugars','sodium','fiber']).sort_values(by='rating', ascending=False) 
    elif age < 14:
        #According to the guideline, it says 'sodium intake should be EVEN LESS if younger than 14.'... but they can be 10% of calories worth of fat and sugar
        return df.nsmallest(n=n, columns=['sodium']).sort_values(by='rating', ascending=False)
    else:
        #Since all the cereals meet the guideline, we can just return the top 5 rating cereal based on consumer reports.
        return df.sort_values(by='rating', ascending=False).head(n)
```

### For Infants, the top 5 recommended cereal are:

```python
# Accomodates for infants less than 2 years old... where the guidelines states to avoid added sugar for infants and toddlers...and sodium... also, avoid too much fiber (extra research)
infants = top_n_cereal(df, 5, 1)
infants
```

```python
# horizontal bar chart of infants recommended cereals
infants.plot.barh(x='name', y='rating', figsize=(10,5))
plt.ylabel('Cereal brand')
plt.xlabel('Rating')
plt.title('Recommended Cereals for Infants')
plt.show()
```

### For Ages 2-14, the top 5 recommended cereal are:

```python
# For childrens >= 2 and less than 14, they can have some sugar (less than 10% of calories per day) and fats (less than 10% of calories per day)
young_teens_and_under = top_n_cereal(df, 5, 7)
young_teens_and_under
```

```python
# horizontal bar chart of young_teens_and_under recommended cereals
young_teens_and_under.plot.barh(x='name', y='rating', figsize=(10,5))
plt.ylabel('Cereal brand')
plt.xlabel('Rating')
plt.title('Recommended Cereals for Age 2-14')
plt.show()
```

### For teens and adults (Age 14+), the top 5 recommended cereal are:

```python
# Because all of the cereal meets the requirements of the 4 guidelines that includes quantitative recommendations, it's best to chose based on ratings
teens_and_adults = top_n_cereal(df, 5, 19)
teens_and_adults
```

```python
# horizontal bar chart of teens_and_adult recommended cereals
teens_and_adults.plot.barh(x='name', y='rating', figsize=(10,5))
plt.ylabel('Cereal brand')
plt.xlabel('Rating')
plt.title('Recommended Cereals for Teens and Adults')
plt.show()
```

### To view the Pearson Correlation between the independent variables and the rating:

With the Pearson's Correlation, we can see that rating is correlated with a little over half of the independent features:

### Additional visualization with BoxPlot

```python
# We can visually see what the driving force of rating is, with either a box plot or just a scatter plot. In this case, the less amount of fats the cereal has, the higher the rating
import seaborn as sns
sns.boxplot(data=df, x=df['fat'],y=df['rating'])
```

```python
# Similarly, the less amount of sugar the cereal has, the higher the rating
sns.boxplot(data=df, x=df['sugars'],y=df['rating'])
```

```python
# The less the calories, the higher the rating
sns.boxplot(data=df, x=df['calories'],y=df['rating'])
```

It seems that the ratings was most likely based on health rather than popularity, which is why it's a good indicator of recommendation from my function.

```python
# This is just personal preferences, if someone wanted like no fat, no sugar, no sodium... but perhaps don't care about the other independent features... 
# based on ratings, we can just mess around with this code:
custom_columns = ['fat','sugars','sodium']
def preference_cereal(df, n, target):
    return df.nsmallest(n=n, columns=custom_columns).sort_values(by=target, ascending=False) 
# plot the dataframe of top 5:
preference_cereal(df,5,'rating').plot.barh(x='name', y='rating', figsize=(10,5))
plt.ylabel('Cereal brand')
plt.xlabel('rating')
plt.title('Recommended Cereals with Custom Columns')
plt.show()
```

```python
preference_cereal(df,5,'rating')
```
